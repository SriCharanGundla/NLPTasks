```python
def gender_features(word):
    return {'last_letter':word[-1]}  #returns the last character
```

```python
gender_features('Charan')
```

```python
from nltk.corpus import names
```

```python
#name and gender list
labeled_names = ([(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names) #shuffles the list
```

```python
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]
```

```python
featuresets
```

```python
len(featuresets) #number of records
```

```python
train_set, test_set = featuresets[500:], featuresets[:500] 
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set) #trains the dataset
```

```python
classifier.classify(gender_features('David'))
```

```python
classifier.classify(gender_features('Charan'))
```

```python
classifier.classify(gender_features('Obama'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
